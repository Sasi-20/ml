```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('winequality-red.csv')
N = df.shape[0]
X = df.drop(columns=["volatile acidity"])
y = df["volatile acidity"]
y = y.values.reshape(-1, 1)
print("X shape:", X.shape)
print("y shape:", y.shape)
split_ratio = 0.8
split_index = int(N * split_ratio)
X_train = X[:split_index]
y_train = y[:split_index]
X_test = X[split_index:]
y_test = y[split_index:]
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
X_train_transpose = np.transpose(X_train)
w = np.linalg.inv(X_train_transpose.dot(X_train)).dot(X_train_transpose).dot(y_train)
y_pred = X_test.dot(w)
mse = np.mean((y_test - y_pred) ** 2)
print("Mean Squared Error (MSE) on test data:", mse)
correct_predictions = np.sum(y_test == y_pred)
total_predictions = len(y_test)
accuracy = correct_predictions  / total_predictions
print("accuracy:",accuracy)

plt.scatter(y_test, y_pred)
plt.xlabel("target Data(Ground Truth)")
plt.ylabel("Predictions")
plt.title("Predictions vs. target Data")
plt.show()


```
